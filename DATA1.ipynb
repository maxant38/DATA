```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv ('synthese-fra.csv ', sep =',')
df["dateUpdate"] = pd.to_datetime(df["date"]) # necessaire pour avoir un truc lisible en axis
df . head (50)


```

```python
x=df.dateUpdate
y=df.total_cas_confirmes
plt.plot(x,y)
plt.title("volution du nombre de cas confirms pendant un an", fontsize=18, y=1.1)
plt.xlabel("date", fontsize=12, labelpad=20)
plt.ylabel("nombre de cas confirms", fontsize=12, labelpad=20)
plt.tick_params(labelsize=10, pad=10, color='red');
plt.show()
```

```python
fig, axs = plt.subplots(3,1,sharex=True, figsize=(15,10))
yOne=df["patients_reanimation"]
yTwo=df["total_deces_hopital"]
yThree=df["total_cas_confirmes"]
x = df["dateUpdate"]
fig.suptitle('Comparaison patient en ranimation - total dcs en hpital - total de cas')
axs[0].plot(x, yOne, 'tab:orange')
axs[1].plot(x, yTwo, 'tab:green')
axs[2].plot(x, yThree, 'tab:blue')
plt.xlabel("date", fontsize=20)
axs[0].set(ylabel='patient_ranimation')
axs[1].set(ylabel='total_deces_hopital')
axs[2].set(ylabel='total_cas_confirmes')
plt.show()
```
